#📌 Extracción

# Sección nueva

```python
import pandas as pd

# Cargar JSON local en DataFrame
file_path = "/content/TelecomX_Data.json"
data = pd.read_json(file_path)

# Ver primeras filas
print("Datos originales:")
display(data.head())

# Información general
print("\nInformación del DataFrame:")
data.info()
```

#🔧 Transformación

```python
# =========================
# 🔧 TRANSFORMACIÓN CORREGIDA
# =========================
import pandas as pd

# Convertir columnas con dict o lista a string para evitar errores
for col in data.columns:
    if data[col].apply(type).eq(dict).any() or data[col].apply(type).eq(list).any():
        data[col] = data[col].astype(str)

# Eliminar duplicados
data.drop_duplicates(inplace=True)

# Revisar y contar valores nulos
print("\nValores nulos por columna:")
print(data.isnull().sum())

# Llenar valores nulos
for col in data.select_dtypes(include='number').columns:
    data[col].fillna(0, inplace=True)

for col in data.select_dtypes(include='object').columns:
    data[col].fillna('Desconocido', inplace=True)

# Ver primeras filas después de la limpieza
print("\nDatos después de la transformación:")
display(data.head())

```

#📊 Carga y análisis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Normalizar nombres de columnas (minúsculas y sin espacios)
data.rename(columns=lambda x: x.strip().lower(), inplace=True)

# Revisar nombres de columnas
print("Columnas en el DataFrame:")
print(data.columns.tolist())

# -------------------------
# Distribución de Churn
# -------------------------
if 'churn' in data.columns:
    # Crear una columna de proporciones si es que no existe
    churn_counts = data['churn'].value_counts(normalize=True)
    print("\nDistribución de Churn:")
    print(churn_counts)

    # Gráfico de barras de Churn
    plt.figure(figsize=(6,4))
    sns.countplot(x='churn', data=data)
    plt.title('Distribución de Churn')
    plt.show()

    # -------------------------
    # Churn por tipo de plan
    # -------------------------
    if 'plan' in data.columns:
        plt.figure(figsize=(8,5))
        sns.countplot(x='plan', hue='churn', data=data)
        plt.title('Churn por tipo de plan')
        plt.show()

    # -------------------------
    # Edad de clientes vs Churn
    # -------------------------
    if 'edad' in data.columns:
        plt.figure(figsize=(8,5))
        sns.boxplot(x='churn', y='edad', data=data)
        plt.title('Distribución de edad según Churn')
        plt.show()
else:
    print("La columna 'churn' no existe en el DataFrame. Revisa el nombre exacto.")

# -------------------------
# Correlaciones
# -------------------------
# Seleccionar solo columnas numéricas
numericas = data.select_dtypes(include='number')

if not numericas.empty:
    plt.figure(figsize=(10,8))
    sns.heatmap(numericas.corr(), annot=True, cmap='coolwarm')
    plt.title('Mapa de correlaciones (solo columnas numéricas)')
    plt.show()
else:
    print("No hay columnas numéricas para mostrar en el heatmap.")

# -------------------------
# Guardar CSV limpio
# -------------------------
data.to_csv("/content/churn_limpio.csv", index=False)
print("\nCSV limpio guardado como 'churn_limpio.csv'.")
```
